# 3. Exploratory Data Analysis(EDA)

### 3.1 Demographic Analysis:

## ðŸ‘‡ Categorising Age Distribution

```python
# Demographic Analysis
age_distribution = data['Age'].value_counts()
gender_distribution = data['Gender'].value_counts()
blood_type_distribution = data['Blood Type'].value_counts()
# Plotting Age Distribution
plt.figure(figsize=(15, 6))
plt.bar(age_distribution.index, age_distribution.values, color='skyblue')
plt.xlabel('Age',weight = 'bold', fontsize = 15)
plt.ylabel('Number of Patients',weight = 'bold',fontsize = 10)
plt.title('Age Distribution of Patients',weight = 'bold',fontsize = 20)
plt.grid(True)
plt.show()
```

## ðŸ‘‡ Percentage of Gender Distribution of Patients

```python
# Plotting Gender Distribution
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution of Patients')
plt.show()
```

- The above pie chart illustrates the gender distribution among patients:
                                                                      Females:- 50.7%
                                                                      Males :- 49.3%
                                                                          

## ðŸ‘‡ Categorising Blood Type Distribution

```python
# Plotting Blood Type Distribution
plt.figure(figsize=(8, 6))
blood_type_distribution.plot(kind='bar', color='lightgreen', ylim= (1200,1300))
ax = blood_type_distribution.plot(kind='bar', stacked=False, edgecolor='k')

# Add bar values on top of bars
for p in ax.patches:
    ax.annotate(np.round(p.get_height(), decimals=2),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5),
                textcoords='offset points')
plt.xlabel('Blood Type',weight = 'bold', fontsize = 15)
plt.ylabel('Number of Patients',weight = 'bold', fontsize = 15)
plt.title('Blood Type Distribution of Patients',weight = 'bold',fontsize = 20)
plt.show()
```

### The Above bar chart displays the Blood type distribution for 10000 patients in the Dataset:
- AB- as the most prevalent blood type patients and A- as the least prevalent blood type patients.

## ðŸ‘‡ Created a heatmap of admission types based on age

```python
# Create a heatmap of admission types based on age
age_admission_trends = data.groupby(['Age', 'Admission Type'])['Name'].count().unstack()
plt.figure(figsize=(20, 20))  
sns.heatmap(age_admission_trends, cmap="Reds", annot=True, fmt='d')  # Create the heatmap
plt.xlabel('Admission Type',weight = 'bold', fontsize = 15)  
plt.ylabel('Age',weight = 'bold', fontsize = 15)
plt.title('Heatmap of Admission Types based on Age',weight = 'bold', fontsize = 20)
plt.xticks(rotation=0)  
plt.yticks(rotation=0)  
plt.show()  
```

## ðŸ‘‡ Group data by Gender and Admission Type and created a bar plot

```python
gender_admission_trends = data.groupby(['Gender', 'Admission Type'])['Name'].count().unstack()

ax = gender_admission_trends.plot(kind='bar', stacked=False, edgecolor='k')  # Create the bar plot

# Add bar values on top of bars
for p in ax.patches:
    ax.annotate(np.round(p.get_height(), decimals=2),  # Annotate the bar height
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Position of the annotation
                ha='center', va='center', xytext=(0, 5),  # Alignment and offset of the annotation
                textcoords='offset points')  # Annotation text coordinates

plt.xlabel('Gender',weight = 'bold', fontsize = 15)
plt.ylabel('Number of Patients',weight = 'bold', fontsize = 15)
plt.title('Trends in Admission Types based on Gender',weight = 'bold',fontsize = 18)
plt.xticks(rotation=0)
plt.legend(title='Admission Type')
plt.tight_layout()
plt.show()
```

- The Above bar chart displays in each and every Admission type Females are higher compared with Males.

## ðŸ‘‡ Plotting the average billing amount by medical condition

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12, 6))

# Group data by Medical Condition and calculate the average billing amount
avg_cost_by_condition = data.groupby('Medical Condition')['Billing Amount'].mean()

# Plot the data with green circles and dashed lines
plt.plot(avg_cost_by_condition, "g*", markersize=10)  

# Annotate each point with its corresponding value
for condition, cost in avg_cost_by_condition.items():
    plt.annotate(f"{cost:.2f} â‚¹", (condition, cost), textcoords="offset points", xytext=(0,6), ha='center')

plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.xlabel('Medical condition',weight = 'bold', fontsize = 15)
plt.ylabel('Average Billing Amount',weight = 'bold', fontsize = 15)  # Set the label for the y-axis
plt.title('Average Billing Amount by Medical Condition',weight = 'bold', fontsize = 20)  # Set the title of the plot
plt.grid(True)  # Add gridlines to the plot
plt.show()  # Display the plot
```

- This plot visualizes the average billing amount for different medical conditions like Arthritis, Asthma, Cancer, Diabetes, Hypertension and Obesity.
- Diabetes has the highest average billing amount, while Arthritis has the lowest.
- The plot highlights variations in billing amounts across various medical conditions, aiding in understanding healthcare expenditure patterns.

## ðŸ‘‡ Plotting Average Billing Amount Vs Admission Type

```python
plt.figure(figsize=(12, 6))
avg_cost_by_admission_type = data.groupby('Admission Type')['Billing Amount'].mean()
plt.plot(avg_cost_by_admission_type, "bo",markersize=4)  # Plot the data with blue circles and dashed lines

# Annotate each point with its corresponding value
for x, y in zip(avg_cost_by_admission_type.index, avg_cost_by_admission_type):
    plt.text(x, y, f'{y:.2f} â‚¹', ha='center', va='bottom')

plt.title("Average Billing Amount by Admission Type",weight = 'bold', fontsize = 20)  # Set the title of the plot
plt.ylabel("Average Billing Amount",weight = 'bold', fontsize = 15)  # Set the label for the y-axis
plt.xlabel('Admission Type',weight = 'bold', fontsize = 15)  # Set the label for the x-axis
plt.grid(True)  # Add gridlines to the plot
plt.show()  # Display the plot
```

- This plot depicts the average billing amount across different admission types.
- Urgent admission type exhibits the highest average billing amount, while emergency admission type displays the lowest.
- Understanding variations in billing amounts by admission type aids in healthcare resource allocation and financial planning.

## ðŸ‘‡ Created a histogram of the Billing Amount Vs No. fo patients

```python
plt.figure(figsize=(12, 6))
hist = data['Billing Amount'].plot.hist(bins=20, edgecolor='k', ylim=(400, 600))  # Adjust bin count and ylim as needed
plt.title("Distribution of Billing Amounts",weight = 'bold', fontsize = 20)
plt.xlabel("Billing Amount",weight = 'bold', fontsize = 15)
plt.ylabel("Number of Patients",weight = 'bold', fontsize = 15)
plt.grid(True)

# Annotate each bar with its corresponding count
for rect in hist.patches:
    plt.text(rect.get_x() + rect.get_width() / 2, rect.get_height(), int(rect.get_height()),
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
